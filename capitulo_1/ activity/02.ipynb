Consultar otros datos estadísticos, conteo, mediana, valores mínimo y máximo.

```python
import pandas as pd
df = pd.read_csv( '../database/inmuebles_bogota.csv' )
col = {
    'Tipo': 'type',
    'Descripcion':'description',
    'Habitaciones':'bedrooms',
    'Baños':'bathrooms',
    'Área':'area',
    'Barrio':'mayoralty',
    'UPZ':'suburb',
    'Valor':'value'
}
df = df.rename( columns = col )
```

La funcion describe retotna una desscripcion con los vlores count, mean, stad, min, max, y un por sentaje de 25% del 25 al 75, la funcion sort_values nos ayuda a organizar los valorea dependieno a los valores que le demos a mostrar.

```python
df.groupby(['mayoralty'])['area'].describe().sort_values(by='mean', ascending=False)
```
